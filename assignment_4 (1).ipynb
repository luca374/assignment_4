```python
import requests
import csv
import pandas as pd
from bs4 import BeautifulSoup

header = ['page #','brand','url','product','price']
data = []
# Display and store away 2 pages of scrapped data from startupbeat.hkej.com
for i in range(1,4):
    quote_page = requests.get('https://catalog.hkstore.com/catalog_tc_hk/men?p='+str(i)) 
    print("\n***** Page " + str(i) +" in action *****")
    soup = BeautifulSoup(quote_page.content,'html.parser')
    # for item in soup.find_all("li", attrs={"class":"item product product-item"}):
    for item in soup.find_all("li", attrs={"class":"item product product-item"}):
         page_no = str(i)
         brand = item.find("div",attrs={"class":"product-item-brand"}).span.get_text()
         url = item.find("strong",attrs={"class":"product name product-item-name"}).a.get('href')
         product = item.find("a",attrs={"class":"product-item-link"}).get_text().strip()
         price = item.find("span",attrs={"class":"price"}).get_text().strip()
         data.append((page_no, brand, url, product, price))
        
df = pd.DataFrame(data,
   columns = header
)
df.to_csv('eshop_catalog.csv', sep='\t', encoding='utf-8')
df
```
